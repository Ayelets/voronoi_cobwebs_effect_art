First we upload a jpg file

```python
from google.colab import files

uploaded = files.upload()
```

Now load the file

```python
from PIL import Image, ImageFilter
import numpy as np
import matplotlib.pyplot as plt
import cv2

from PIL import Image 
image_file = Image.open(list(uploaded.keys())[0]) # open colour image
image_file = image_file.convert('L') # Convert to grayscale
```

Convert to black and white

```python
threshold = 160
image_file = image_file.point(lambda i: 0 if i<threshold else 255)
```

Collect all black points in a list

```python
[x_b,y_b]=np.where(np.array(image_file)==0)
blacklist=list(zip(y_b,-x_b))
```

Select a random set from the black points

```python
import random
num_of_centers=8300
centers=random.sample(blacklist,num_of_centers)
```

Run Vornoi algorithm

```python
from scipy.spatial import Voronoi, voronoi_plot_2d
vor = Voronoi(centers)
voronoi_plot_2d(vor,show_points=False,show_vertices=False)
```
